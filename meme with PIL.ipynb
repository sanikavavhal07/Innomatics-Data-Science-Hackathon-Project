```python
import matplotlib.pyplot as plt
from PIL import Image,ImageDraw,ImageFont
img=Image.open('46hhvr.jpg')

print("image size:",img.size)
print("image mode:",img.mode)
plt.imshow(img)
```

```python
draw=ImageDraw.Draw(img)

TEXT1="PEOPLE WITH COLDPLAY TICKETS"
TEXT2="OTHERS WITH NO TICKET"
TEXT3="ME DOING GENAI TASK 4"
font= ImageFont.truetype('Roboto-Black.ttf',30)

draw.text((200,100),text=TEXT1, fill=(0,0,0),font=font, align="centre")
draw.text((20,300),text=TEXT2, fill=(0,0,0),font=font, align="centre")
draw.text((300,1000),text=TEXT3, fill=(0,0,0),font=font, align="centre")
plt.imshow(img)
```
